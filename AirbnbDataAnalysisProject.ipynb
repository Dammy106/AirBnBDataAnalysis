**Data Analysis**

```python
#Changing scope to remove that outlier
data = df[df['price']<1500]
```

```python
##Identifying outliers
sns.boxplot(data=data,x='price')
#1 huge outlier of 100000
```

```python
##Price Distributions
```

```python
plt.figure(figsize=(8,6))
plt.title("Price Distribution")
plt.ylabel('Frequency')
plt.xlabel('Price')
sns.histplot(data=data,x='price',bins=100)
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.title("Availability_365 Distribution")
plt.ylabel('Frequency')
plt.xlabel('Availability_365')
sns.histplot(data=data,x='availability_365')
plt.show()
```

```python
data.groupby(by='neighbourhood_group')['price'].mean()
```

**Feature Engineering**

```python
#Price per bed
data['price per bed'] = data['price']/data['beds']
data.head()
```

```python
data.groupby(by='neighbourhood_group')['price per bed'].mean()
```

```python
sns.barplot(data=data,x='neighbourhood_group',y='price',hue='room_type')
```

```python
#number of review and price relationship
plt.title("Number_of_Reviews and Price relationship per Neighbourhood_Group")
sns.scatterplot(data=data,x='price',y='number_of_reviews',hue='neighbourhood_group')
```

```python
##
sns.pairplot(data=data,vars=['price','minimum_nights','number_of_reviews','availability_365'],hue='room_type')
```
